# Data Loading and Setup
## House Prices in Grand Tunis - Data Mining Project

This notebook contains the initial data loading and setup steps for analyzing property prices in Grand Tunis, Tunisia.

## Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Data Loading

Loading the Property Prices in Tunisia dataset from the data directory.

```python
# Load the dataset
df = pd.read_csv('../data/raw/source_1/Property-Prices-in-Tunisia.csv')
print(f"Dataset shape: {df.shape}")
print(f"Column names: {df.columns.tolist()}")
display(df.head())
```

## Initial Data Inspection

Let's examine the basic structure and characteristics of our dataset.

```python
# Basic information about the dataset
print("Dataset Info:")
display(df.info())

print("\nDataset Description:")
display(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nUnique values in categorical columns:")
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f"{col}: {df[col].nunique()} unique values")
    print(f"Values: {df[col].unique()}")
    print()
```
